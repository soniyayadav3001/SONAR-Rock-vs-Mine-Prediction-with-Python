```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas Dataframe
sonar_data = pd.read_csv('sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()  #describe --> statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
# separating data and Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

<h1>Training and Test data</h1>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

<h1>Model Training --> Logistic Regression</h1>

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

<h1>Model Selection and Hyperparameter Tuning</h1>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# Define models and their hyperparameters
models = {
    'Logistic Regression': (LogisticRegression(), {'C': [0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']}),
    'Random Forest': (RandomForestClassifier(), {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}),
    'SVM': (SVC(), {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']})
}
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

# Perform GridSearchCV for each model
best_models = {}
for model_name, (model, params) in models.items():
    grid_search = GridSearchCV(model, params, cv=5, scoring='accuracy')
    grid_search.fit(X_train_scaled, Y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best cross-validation accuracy for {model_name}: {grid_search.best_score_:.4f}\n")
```

<h1>Model Evaluation Metrics

```python
from sklearn.metrics import classification_report, roc_auc_score, roc_curve

# Evaluate the best model (e.g., Logistic Regression)
best_model = best_models['Logistic Regression']
X_test_scaled = scaler.transform(X_test)
Y_pred = best_model.predict(X_test_scaled)



# Classification report
print("Classification Report:")
print(classification_report(Y_test, Y_pred))

# ROC-AUC score
Y_pred_proba = best_model.predict_proba(X_test_scaled)[:, 1]
roc_auc = roc_auc_score(Y_test, Y_pred_proba)
print(f"ROC-AUC Score: {roc_auc:.4f}")

# ROC-AUC score and ROC curve
# Convert Y_test to binary format (0 for 'R', 1 for 'M')
Y_test_binary = Y_test.map({'R': 0, 'M': 1})
# Or use LabelEncoder for more complex cases:
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# Y_test_binary = le.fit_transform(Y_test)

Y_pred_proba = best_model.predict_proba(X_test_scaled)[:, 1]
roc_auc = roc_auc_score(Y_test_binary, Y_pred_proba)  # Use binary Y_test
print(f"ROC-AUC Score: {roc_auc:.4f}")

fpr, tpr, thresholds = roc_curve(Y_test_binary, Y_pred_proba)  # Use binary Y_test
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# Assuming 'scaler' is the StandardScaler instance used earlier
X_scaled = scaler.fit_transform(X)  # Scaling the entire dataset X

# Perform 5-fold cross-validation on the best model
cv_scores = cross_val_score(best_model, X_scaled, Y, cv=5, scoring='accuracy')

print("Cross-Validation Scores:", cv_scores)
print(f"Mean Cross-Validation Accuracy: {cv_scores.mean():.4f}")
```
